```python
import pandas as pd
from matplotlib import pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
dataset_1 = pd.read_csv(r'C:\Users\Jaum\Downloads\airtravel.csv', sep=',', index_col='Date', parse_dates=True)
dataset_2= pd.read_csv(r'C:\Users\Jaum\Downloads\co2_emissions.csv', sep=',', index_col='Year', parse_dates=True)
```

```python
dataset_1
```

```python
dataset_1['Passengers'].astype(int)
dataset_1['Passengers'].plot(figsize=(6,8), title='Passageiros nos meses', xlabel='Data', ylabel='Passageiros')
```

```python
decomposition_1 = seasonal_decompose(dataset_1['Passengers'], model='additive')
decomposition_1.plot()
plt.tight_layout()
```

```python
print('Exercicio 1')
print('a) A tendncia  aumentar com o tempo')
print('b) A sazonalidade ocorre num periodo de 1 ano')
print('c) Sim, em julho o nmero de passageiros aumenta, pois  poca de frias\n')
```
